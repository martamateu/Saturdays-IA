```python
import numpy as np
import pandas as pd
pd.set_option('display.max.columns', 100)
# to draw pictures in jupyter notebook
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns
# we don't like warnings
# you can comment the following 2 lines if you'd like to
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('adult.data.csv')
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
data.loc[data['sex'] == 'Female', 'age'].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
germany=data[data['native-country'] == "Germany"]["native-country"].count()  #count german citizens
total=data["native-country"].count()  #count total citizens
percentage=(germany/total)*100    #find percentage
percentage #print percentage
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
#whose value is less than 50K
print("mean of age for those who earn less than 50K:",data[data['salary']=="<=50K"]["age"].mean())
print("standard deviation of age for those who earn less than 50K:",data[data['salary']=="<=50K"]["age"].std())
#whose value is greater than 50K
print("mean of age for those who earn more than 50K:",data[data['salary']==">50K"]["age"].mean())
print("standard deviation of age for those who earn more than 50K:",data[data['salary']==">50K"]["age"].std())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education â€“ Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
array=[ "11th", "9th", "7th-8th", "12th","1st-4th", "10th",  "5th-6th", "Preschool"]
~data.loc[data['salary']==">50K"]["education"].isin(array).all()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
#race by describe()
race_describe=data.groupby("race").describe()

#gender by describe()
sex_describe=data.groupby("sex").describe()

print("race_describe",race_describe,"\n gender_describe",sex_describe)
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
married=["Married-civ-spouse", "Married-spouse-absent","Married-AF-spouse"]
a=data[data['salary']==">50K"]["marital-status"].isin(married).value_counts()
if a[True]>a[False]:
    print("Married")
else:
    print("Bachelors")
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
print("Maximum no. of hours a person works per week :",data["hours-per-week"].max())
no_of_people=data[data["hours-per-week"]==data["hours-per-week"].max()].count().max()
print(no_of_people,"people work",data["hours-per-week"].max(),"hours")
maxhours_maxsalary=(data[data["hours-per-week"]==data["hours-per-week"].max()]["salary"]==">50K").value_counts()
percentage=((maxhours_maxsalary[True])/no_of_people)*100
print("percentage of those who earn a lot (>50K) : ",percentage)
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
for(country,salary),sub_df in data.groupby(["native-country","salary"]):
   print(country, salary, round(sub_df['hours-per-week'].mean(), 2))

```
